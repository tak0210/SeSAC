## 1οΈβƒ£ λ°μ΄ν„° λ΅λ“ λ° μ „μ²λ¦¬

```python
def load_and_preprocess(train_path, test_path):
    print("π“ λ°μ΄ν„° λ΅λ“ λ° μ „μ²λ¦¬...")

    # λ°μ΄ν„° λ΅λ“
    train_df = pd.read_csv(train_path)
    test_df = pd.read_csv(test_path)

    def process_df(df):
        # μ»¬λΌλ… μ •λ¦¬
        if 'Unnamed: 0' in df.columns:
            df = df.drop(columns=['Unnamed: 0'])
        df.columns = [col.replace('train_heat.', '') for col in df.columns]

        # μ‹κ°„ λ³€μ
        df['datetime'] = pd.to_datetime(df['tm'], format='%Y%m%d%H')
        # df['year'] = df['datetime'].dt.year
        df['month'] = df['datetime'].dt.month
        # df['day'] = df['datetime'].dt.day
        df['hour'] = df['datetime'].dt.hour
        df['dayofweek'] = df['datetime'].dt.dayofweek

        # κ²°μΈ΅μΉ μ²λ¦¬
        missing_cols = ['ta', 'wd', 'ws', 'rn_day', 'rn_hr1', 'hm', 'si', 'ta_chi']
        if 'heat_demand' in df.columns:
            missing_cols.append('heat_demand')

        for col in missing_cols:
            if col in df.columns:
                df[col] = df[col].replace(-99, np.nan)

        # wd -9.9 κ²°μΈ΅μΉ μ²λ¦¬
        df.loc[df['wd'] == -9.9, 'wd'] = np.nan

        # μΌμ‚¬λ‰ μ•Όκ°„ μ²λ¦¬
        if 'si' in df.columns:
            night_mask = (df['hour'] < 8) | (df['hour'] > 18)
            df.loc[night_mask & df['si'].isna(), 'si'] = 0

        # μ§€μ‚¬λ³„ λ³΄κ°„
        df = df.sort_values(['branch_id', 'datetime'])
        numeric_cols = df.select_dtypes(include=[np.number]).columns

        for branch in df['branch_id'].unique():
            mask = df['branch_id'] == branch
            df.loc[mask, numeric_cols] = df.loc[mask, numeric_cols].interpolate().fillna(method='ffill').fillna(method='bfill')

        return df

    train_df = process_df(train_df)
    test_df = process_df(test_df)

    print(f"   ν›λ ¨: {train_df.shape}, ν…μ¤νΈ: {test_df.shape}")
    print(f"   κΈ°κ°„: {train_df['datetime'].min()} ~ {test_df['datetime'].max()}")
    print(f"   λΈλμΉ: {sorted(train_df['branch_id'].unique())}")

    return train_df, test_df

train_df, test_df = load_and_preprocess(train_path, test_path)
```

## 2οΈβƒ£ νμƒλ³€μ μƒμ„±

```python
def create_features(df):
    """HDD, wind_chill, μν™ν• μΈμ½”λ”©, λ²”μ£Όν• λ³€μ μƒμ„±"""
    df = df.copy()

    # β­ HDD (μμΉν•)
    if 'ta' in df.columns:
        df['HDD_18'] = np.maximum(18 - df['ta'], 0)
        df['HDD_20'] = np.maximum(20 - df['ta'], 0)

    # β­ wind_chill (μμΉν•)
    if 'ta' in df.columns and 'ws' in df.columns:
        df['wind_chill'] = np.where(
            (df['ta'] <= 10) & (df['ws'] > 0),
            13.12 + 0.6215 * df['ta'] - 11.37 * (df['ws'] ** 0.16) + 0.3965 * df['ta'] * (df['ws'] ** 0.16),
            df['ta']
        )

    # β­ heating_season (λ²”μ£Όν•)
    df['heating_season'] = df['month'].isin([10, 11, 12, 1, 2, 3, 4]).astype(int)

    # μ‹κ°„λ€ λ²”μ£Όν•
    df['is_weekend'] = (df['dayofweek'] >= 5).astype(int)
    df['is_peak_morning'] = ((df['hour'] >= 7) & (df['hour'] <= 9)).astype(int)
    df['is_peak_evening'] = ((df['hour'] >= 18) & (df['hour'] <= 22)).astype(int)
    df['is_night'] = ((df['hour'] >= 23) | (df['hour'] <= 5)).astype(int)

    # ν”Όν¬μ‹κ°„ ν†µν•©
    df['peak_time_category'] = 0
    df.loc[df['is_peak_morning'] == 1, 'peak_time_category'] = 1
    df.loc[df['is_peak_evening'] == 1, 'peak_time_category'] = 2
    df.loc[df['is_night'] == 1, 'peak_time_category'] = 3

    # β­ κΈ°μ¨ λ²”μ£Ό (λ²”μ£Όν•)
    if 'ta' in df.columns:
        df['temp_category'] = pd.cut(df['ta'],
                                   bins=[-np.inf, 0, 10, 20, 30, np.inf],
                                   labels=[0, 1, 2, 3, 4]).astype(int)

    # β­ κ°•μ κ°•λ„ (λ²”μ£Όν•)
    if 'rn_day' in df.columns:
        df['rain_intensity'] = pd.cut(df['rn_day'],
                                   bins=[-1, 0, 1, 5, 10, np.inf],
                                   labels=[0, 1, 2, 3, 4]).astype(int)

    # β­ μν™ν• μΈμ½”λ”© (μ‹κ°„ cos, sin)
    df['hour_sin'] = np.sin(2 * np.pi * df['hour'] / 24)
    df['hour_cos'] = np.cos(2 * np.pi * df['hour'] / 24)
    df['month_sin'] = np.sin(2 * np.pi * df['month'] / 12)
    df['month_cos'] = np.cos(2 * np.pi * df['month'] / 12)
    df['dayofweek_sin'] = np.sin(2 * np.pi * df['dayofweek'] / 7)
    df['dayofweek_cos'] = np.cos(2 * np.pi * df['dayofweek'] / 7)

    # μ‹κ°„ κΈ°λ° νμƒλ³€μ
    df['hour_squared'] = df['hour'] ** 2
    df['month_day_interaction'] = df['month'] * df['datetime'].dt.day
    
    # κΈ°μ¨ κ΄€λ ¨ νμƒλ³€μ
    if 'ta' in df.columns:
        df['ta_squared'] = df['ta'] ** 2
        df['ta_cubed'] = df['ta'] ** 3
    
    # μµλ„μ™€ κΈ°μ¨ μƒνΈμ‘μ©
    if 'hm' in df.columns and 'ta' in df.columns:
        df['hm_ta_interaction'] = df['hm'] * df['ta']

    return df


# νμƒλ³€μ μƒμ„±
train_df = create_features(train_df)
test_df = create_features(test_df)

print(f"β… νμƒλ³€μ μƒμ„± μ™„λ£: {train_df.shape[1]}κ° μ»¬λΌ")
```

## 3οΈβƒ£ μ‹μ¦λ³„-λΈλμΉλ³„ λ°μ΄ν„° λ¶„ν• 

```python
# μ‹μ¦λ³„-λΈλμΉλ³„ λ°μ΄ν„° λ¶„ν• 
def split_by_season_and_branch(df):
    data_splits = {}
    
    branches = sorted(df['branch_id'].unique())
    seasons = [0, 1]  # 0: λΉ„λ‚λ°©μ‹μ¦, 1: λ‚λ°©μ‹μ¦
    season_names = {0: 'λΉ„λ‚λ°©', 1: 'λ‚λ°©'}
    
    print(f"π“ λ°μ΄ν„° λ¶„ν•  μ •λ³΄:")
    print(f"   λΈλμΉ: {len(branches)}κ° - {branches}")
    print(f"   μ‹μ¦: {len(seasons)}κ° - {[season_names[s] for s in seasons]}")
    print(f"   μ΄ μ΅°ν•©: {len(branches) * len(seasons)}κ°")
    
    for season in seasons:
        for branch in branches:
            key = f"{season_names[season]}_{branch}"
            
            # μ‹μ¦κ³Ό λΈλμΉλ΅ ν•„ν„°λ§
            subset = df[(df['heating_season'] == season) & (df['branch_id'] == branch)].copy()
            
            if len(subset) > 0:
                data_splits[key] = subset
                print(f"   {key}: {len(subset):,}κ° λ°μ΄ν„°")
            else:
                print(f"   {key}: λ°μ΄ν„° μ—†μ β οΈ")
    
    return data_splits

# ν›λ ¨ λ° ν…μ¤νΈ λ°μ΄ν„° λ¶„ν• 
train_splits = split_by_season_and_branch(train_df)
test_splits = split_by_season_and_branch(test_df)

print(f"\nβ… ν›λ ¨ λ°μ΄ν„°: {len(train_splits)}κ° λ¶„ν• ")
print(f"β… ν…μ¤νΈ λ°μ΄ν„°: {len(test_splits)}κ° λ¶„ν• ")
```

### π“ Validationμ‹ μ›”λ³„ μ¶”μ¶

```python
def get_temporal_split_indices(df, test_size=0.2, min_val_samples=10):
    """μ‹κ°„ κΈ°λ° λ¶„ν•  μΈλ±μ¤ λ°ν™"""
    
    month_counts = df['month'].value_counts()
    valid_months = month_counts[month_counts >= min_val_samples * 2].index
    
    if len(valid_months) < 3:
        # ν΄λ°±: κΈ°μ΅΄ λ°©μ‹
        split_idx = int(len(df) * (1 - test_size))
        return df.index[:split_idx], df.index[split_idx:]
    
    train_indices = []
    val_indices = []
    
    for month in valid_months:
        month_data = df[df['month'] == month].sort_values('datetime')
        val_size = max(min_val_samples, int(len(month_data) * test_size))
        
        train_indices.extend(month_data.iloc[:-val_size].index)
        val_indices.extend(month_data.iloc[-val_size:].index)
    
    print(f"      π“… μ›”λ³„ λ¶„ν• : {len(valid_months)}κ°μ›”, κ²€μ¦ {len(val_indices)}κ°")
    
    return train_indices, val_indices
```

## 4οΈβƒ£ Catboost λ¨λΈ ν΄λμ¤ μ •μ

```python
class OptimizedCatBoostModel:
    def __init__(self, model_name):
        self.model_name = model_name
        self.model = None
        self.best_params = None
        self.feature_cols = None
        self.study = None
        self.best_score = None
        
        # CatBoost λ²„μ „ ν™•μΈ
        import catboost
        self.catboost_version = catboost.__version__
        
    def define_feature_columns(self, df):
        """νΉμ„± μ»¬λΌ μ •μ"""
        exclude_cols = [
            'tm', 'datetime', 'year', 'heat_demand', 'branch_id'
        ]
        
        self.feature_cols = [col for col in df.columns 
                           if col not in exclude_cols and df[col].dtype in ['int64', 'float64']]
        
        print(f"   π“‹ {self.model_name}: {len(self.feature_cols)}κ° νΉμ„± μ‚¬μ©")
        return self.feature_cols
    
    def get_temporal_split_indices(self, df, test_size=0.2, min_val_samples=10):
        """μ‹κ°„ κΈ°λ° λ¶„ν•  μΈλ±μ¤ λ°ν™"""
        
        month_counts = df['month'].value_counts()
        valid_months = month_counts[month_counts >= min_val_samples * 2].index
        
        if len(valid_months) < 3:
            # ν΄λ°±: κΈ°μ΅΄ λ°©μ‹
            split_idx = int(len(df) * (1 - test_size))
            return df.index[:split_idx], df.index[split_idx:]
        
        train_indices = []
        val_indices = []
        
        for month in valid_months:
            month_data = df[df['month'] == month].sort_values('datetime')
            val_size = max(min_val_samples, int(len(month_data) * test_size))
            
            train_indices.extend(month_data.iloc[:-val_size].index)
            val_indices.extend(month_data.iloc[-val_size:].index)
        
        print(f"      π“… μ›”λ³„ λ¶„ν• : {len(valid_months)}κ°μ›”, κ²€μ¦ {len(val_indices)}κ°")
        
        return train_indices, val_indices
    
    def objective(self, trial, X_train, y_train, X_val, y_val):
        """Optuna λ©μ  ν•¨μ (CatBoost νλΌλ―Έν„°)"""
        # CatBoost ν•μ΄νΌνλΌλ―Έν„° νƒμƒ‰ κ³µκ°„
        params = {
            'iterations': trial.suggest_int('iterations', 100, 1000),
            'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
            'depth': trial.suggest_int('depth', 4, 10),
            'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1, 10, log=True),
            'bagging_temperature': trial.suggest_float('bagging_temperature', 0, 1),
            'random_strength': trial.suggest_float('random_strength', 1e-8, 10, log=True),
            'border_count': trial.suggest_int('border_count', 32, 255),
            'feature_border_type': trial.suggest_categorical('feature_border_type', ['Median', 'Uniform']),
            'loss_function': 'RMSE',
            'random_seed': 42,
            'verbose': False,  # λ΅κ·Έ μ¶λ ¥ μ•ν•¨
            'thread_count': -1
        }
        
        # λ¨λΈ μƒμ„± λ° ν›λ ¨ (early stopping ν¬ν•¨)
        model = CatBoostRegressor(**params)
        
        model.fit(
            X_train, y_train,
            eval_set=(X_val, y_val),
            early_stopping_rounds=50,
            verbose=False
        )
        
        # κ²€μ¦ μμΈ΅
        y_pred = model.predict(X_val)
        rmse = np.sqrt(mean_squared_error(y_val, y_pred))
        
        return rmse
    
    def fit(self, df, target_col='heat_demand', n_trials=20):
        """λ¨λΈ ν›λ ¨ (Optuna μµμ ν™” ν¬ν•¨)"""
        print(f"\nπ” {self.model_name} λ¨λΈ ν›λ ¨ μ‹μ‘...")
        print(f"   π”§ CatBoost λ²„μ „: {self.catboost_version}")
        
        if len(df) < 10:
            print(f"   β οΈ λ°μ΄ν„° λ¶€μ΅± ({len(df)}κ°) - κΈ°λ³Έ λ¨λΈ μ‚¬μ©")
            self._fit_basic_model(df, target_col)
            return
        
        # 1. λ¶„ν•  μΈλ±μ¤ λ¨Όμ € κµ¬ν•κΈ° (datetime μμ„ λ•)
        train_indices, val_indices = self.get_temporal_split_indices(df, test_size=0.2)
        
        # 2. νΉμ„± μ»¬λΌ μ •μ (datetime μ κ±°)
        self.define_feature_columns(df)
        
        # 3. X, y μƒμ„± (datetime μ—†μ)
        X = df[self.feature_cols].copy()
        y = df[target_col].copy()
        
        # 4. μΈλ±μ¤λ΅ λ¶„ν• 
        X_train = X.loc[train_indices]
        y_train = y.loc[train_indices]
        X_val = X.loc[val_indices]
        y_val = y.loc[val_indices]
        
        print(f"      π“ ν›λ ¨: {len(X_train):,}κ°, κ²€μ¦: {len(X_val):,}κ°")
        
        if len(X_train) < 5 or len(X_val) < 2:
            print(f"   β οΈ λ¶„ν•  ν›„ λ°μ΄ν„° λ¶€μ΅± - κΈ°λ³Έ λ¨λΈ μ‚¬μ©")
            self._fit_basic_model(df, target_col)
            return
        
        # Optuna μµμ ν™”
        print(f"   π― ν•μ΄νΌνλΌλ―Έν„° μµμ ν™”: ", end="", flush=True)
        
        try:
            # λ¨λ“  λ΅κ·Έ λ„κΈ°
            optuna.logging.set_verbosity(optuna.logging.ERROR)
            
            self.study = optuna.create_study(
                direction='minimize',
                sampler=TPESampler(seed=42),
                study_name=f"catboost_{self.model_name}"
            )
            
            # μ§„ν–‰μƒν™© ν‘μ‹λ¥Ό μ„ν• μ½λ°±
            def progress_callback(study, trial):
                print(f"\r   π― ν•μ΄νΌνλΌλ―Έν„° μµμ ν™”: {len(study.trials)}/{n_trials} (Best RMSE: {study.best_value:.4f})", end="", flush=True)
            
            # μµμ ν™” μ‹¤ν–‰
            self.study.optimize(
                lambda trial: self.objective(trial, X_train, y_train, X_val, y_val),
                n_trials=n_trials,
                callbacks=[progress_callback]
            )
            
            print()  # μ¤„λ°”κΏ
            
            # μ•μ „ν• best_value μ ‘κ·Ό
            if len(self.study.trials) > 0 and self.study.best_trial is not None:
                self.best_score = self.study.best_value
                self.best_params = self.study.best_params.copy()
                
                # κΈ°λ³Έ νλΌλ―Έν„° μ¶”κ°€
                self.best_params.update({
                    'loss_function': 'RMSE',
                    'random_seed': 42,
                    'verbose': False,
                    'thread_count': -1
                })
                
                # μ „μ²΄ λ°μ΄ν„°λ΅ μµμΆ… ν›λ ¨
                self.model = CatBoostRegressor(**self.best_params)
                
                self.model.fit(
                    X, y,
                    eval_set=(X_val, y_val),
                    early_stopping_rounds=100,
                    verbose=False
                )
                
                # μ„±λ¥ μ •λ³΄
                val_pred = self.model.predict(X_val)
                val_rmse = np.sqrt(mean_squared_error(y_val, val_pred))
                
                print(f"   π“ μµμ ν™” μ™„λ£: Best RMSE = {self.best_score:.4f}")
                print(f"   π“ κ²€μ¦ RMSE = {val_rmse:.4f}")
                print(f"   π† μµμ  νλΌλ―Έν„°: lr={self.best_params.get('learning_rate', 0.1):.3f}, depth={self.best_params.get('depth', 6)}")
                
            else:
                print(f"   β οΈ μµμ ν™” μ‹¤ν¨: μ ν¨ν• trial μ—†μ - κΈ°λ³Έ λ¨λΈ μ‚¬μ©")
                self._fit_basic_model(df, target_col)
                
        except Exception as e:
            print(f"\n   β οΈ μµμ ν™” μ‹¤ν¨: {str(e)[:30]}... - κΈ°λ³Έ λ¨λΈ μ‚¬μ©")
            self._fit_basic_model(df, target_col)
    
    def _fit_basic_model(self, df, target_col):
        """κΈ°λ³Έ λ¨λΈ ν›λ ¨ (μµμ ν™” μ‹¤ν¨ μ‹ μ‚¬μ©)"""
        if not hasattr(self, 'feature_cols') or self.feature_cols is None:
            self.define_feature_columns(df)
        
        X = df[self.feature_cols].copy()
        y = df[target_col].copy()
        
        # κΈ°λ³Έ CatBoost νλΌλ―Έν„°
        self.model = CatBoostRegressor(
            iterations=500,
            learning_rate=0.1,
            depth=6,
            l2_leaf_reg=3,
            loss_function='RMSE',
            random_seed=42,
            verbose=False,
            thread_count=-1
        )
        
        self.model.fit(X, y, verbose=False)
        
        self.best_score = None
        self.best_params = None
        print(f"   π”§ κΈ°λ³Έ λ¨λΈ ν›λ ¨ μ™„λ£")
    
    def predict(self, df):
        """μμΈ΅"""
        if self.model is None:
            return np.full(len(df), 0)
        
        # λ™μΌν• νΉμ„± μ»¬λΌ μ‚¬μ©
        X = df[self.feature_cols].copy()
        
        # μμΈ΅
        predictions = self.model.predict(X)
        
        # μμ κ°’ μ²λ¦¬
        predictions = np.maximum(predictions, 0)
        
        return predictions

print("β… CatBoost λ¨λΈ ν΄λμ¤ μ •μ μ™„λ£")
```

## 5οΈβƒ£ 38κ° λ¨λΈ ν›λ ¨

```python
# 38κ° λ¨λΈ ν›λ ¨
print("π€ 38κ° CatBoost λ¨λΈ ν›λ ¨ μ‹μ‘!")
print("=" * 60)

models = {}
training_results = {}
n_trials_per_model = 20  # μ‹κ°„ λ‹¨μ¶•μ„ μ„ν•΄ 20μΌλ΅ μ΅°μ •

start_time = datetime.now()

# μ„±κ³µ/μ‹¤ν¨ μΉ΄μ΄ν„°
success_count = 0
basic_count = 0
failed_count = 0

# λ¨λ“  ν›λ ¨ λ°μ΄ν„° λ¶„ν• μ— λ€ν•΄ λ¨λΈ ν›λ ¨
for i, (model_key, train_data) in enumerate(train_splits.items(), 1):
    print(f"\n[{i:2d}/{len(train_splits)}] π”¥ {model_key}")
    print(f"         π“ λ°μ΄ν„°: {len(train_data):,}κ°")
    
    # λ¨λΈ μƒμ„± λ° ν›λ ¨
    model = OptimizedCatBoostModel(model_key)
    
    try:
        model_start = datetime.now()
        model.fit(train_data, n_trials=n_trials_per_model)
        model_time = (datetime.now() - model_start).total_seconds()
        
        models[model_key] = model
        
        # κ²°κ³Ό λ¶„λ¥
        if model.best_score is not None:
            success_count += 1
            status = "β… μµμ ν™”"
            score_text = f"RMSE: {model.best_score:.3f}"
        else:
            basic_count += 1
            status = "π”§ κΈ°λ³Έλ¨λΈ"
            score_text = "κΈ°λ³ΈνλΌλ―Έν„°"
        
        # μ•μ „ν• κ²°κ³Ό μ €μ¥
        training_results[model_key] = {
            'data_size': len(train_data),
            'training_time': model_time,
            'best_score': model.best_score,
            'best_params': model.best_params,
            'optimization_success': model.best_score is not None
        }
        
        print(f"         {status} | {score_text} | β±οΈ {model_time:.1f}μ΄")
        
    except Exception as e:
        print(f"         β ν›λ ¨ μ‹¤ν¨: {str(e)[:30]}...")
        failed_count += 1
        
        # μ™„μ „ κΈ°λ³Έ λ¨λΈλ΅ λ€μ²΄
        try:
            basic_model = OptimizedLightGBMModel(model_key)
            basic_model._fit_basic_model(train_data, 'heat_demand')
            models[model_key] = basic_model
            
            training_results[model_key] = {
                'data_size': len(train_data),
                'training_time': 0,
                'best_score': None,
                'best_params': None,
                'optimization_success': False,
                'error': str(e)[:50]
            }
            print(f"         π”§ κΈ°λ³Έ λ¨λΈλ΅ λ€μ²΄ μ™„λ£")
            
        except Exception as e2:
            print(f"         β κΈ°λ³Έ λ¨λΈ μƒμ„±λ„ μ‹¤ν¨")
            # λ”λ―Έ λ¨λΈ μ €μ¥
            dummy_model = OptimizedLightGBMModel(model_key)
            dummy_model.model = None
            models[model_key] = dummy_model
            
            training_results[model_key] = {
                'data_size': len(train_data),
                'training_time': 0,
                'best_score': None,
                'best_params': None,
                'optimization_success': False,
                'error': f"Complete failure: {str(e2)[:30]}"
            }

total_time = (datetime.now() - start_time).total_seconds()

print(f"\n" + "=" * 60)
print(f"π‰ λ¨λ“  λ¨λΈ ν›λ ¨ μ™„λ£!")
print(f"β±οΈ  μ΄ μ†μ” μ‹κ°„: {total_time/60:.1f}λ¶„ (ν‰κ·  {total_time/len(train_splits):.1f}μ΄/λ¨λΈ)")
print(f"π“ ν›λ ¨ κ²°κ³Ό:")
print(f"   β… μµμ ν™” μ„±κ³µ: {success_count}κ°")
print(f"   π”§ κΈ°λ³Έ λ¨λΈ: {basic_count}κ°") 
print(f"   β μ™„μ „ μ‹¤ν¨: {failed_count}κ°")
print(f"   π“ μ „μ²΄ μ„±κ³µλ¥ : {(success_count + basic_count)/len(train_splits)*100:.1f}%")

# μµκ³  μ„±λ¥ λ¨λΈ μ°ΎκΈ°
successful_models = {k: v for k, v in training_results.items() 
                    if v.get('optimization_success', False)}

if successful_models:
    best_model = min(successful_models.items(), key=lambda x: x[1]['best_score'])
    print(f"π† μµκ³  μ„±λ¥: {best_model[0]} (RMSE: {best_model[1]['best_score']:.4f})")
    
    # μ‹μ¦λ³„ ν‰κ·  μ„±λ¥
    season_scores = {'λ‚λ°©': [], 'λΉ„λ‚λ°©': []}
    for model_name, result in successful_models.items():
        season = model_name.split('_')[0]
        if season in season_scores:
            season_scores[season].append(result['best_score'])
    
    print(f"π“ μ‹μ¦λ³„ ν‰κ·  RMSE:")
    for season, scores in season_scores.items():
        if scores:
            print(f"   {season}μ‹μ¦: {np.mean(scores):.4f} ({len(scores)}κ° λ¨λΈ)")

print("=" * 60)
```

## 8οΈβƒ£ μμΈ΅ κ²°κ³Ό ν†µν•©

```python
# μμΈ΅ κ²°κ³Όλ¥Ό μ›λ³Έ test_df μμ„μ— λ§κ² ν†µν•©
print("π”„ μμΈ΅ κ²°κ³Ό ν†µν•© μ¤‘...")

# κ²°κ³Όλ¥Ό μ €μ¥ν•  λ°°μ—΄ μ΄κΈ°ν™”
final_predictions = np.zeros(len(test_df))
prediction_counts = np.zeros(len(test_df))  # κ° μΈλ±μ¤λ³„ μμΈ΅ νμ μ¶”μ 

# κ° μμΈ΅ κ²°κ³Όλ¥Ό ν•΄λ‹Ή μΈλ±μ¤μ— ν• λ‹Ή
for model_key, pred_info in predictions.items():
    test_data = pred_info['data']
    pred_values = pred_info['predictions']
    
    # μ›λ³Έ test_dfμ—μ„ ν•΄λ‹Ή λ°μ΄ν„°μ μΈλ±μ¤ μ°ΎκΈ°
    season, branch = model_key.split('_')
    season_num = 1 if season == 'λ‚λ°©' else 0
    
    # μ΅°κ±΄μ— λ§λ” μΈλ±μ¤ μ°ΎκΈ°
    mask = (test_df['heating_season'] == season_num) & (test_df['branch_id'] == branch)
    indices = test_df[mask].index.tolist()
    
    print(f"π“ {model_key}: {len(indices)}κ° μΈλ±μ¤μ— ν• λ‹Ή")
    
    # μμΈ΅κ°’ ν• λ‹Ή (μΈλ±μ¤ κ°μμ™€ μμΈ΅κ°’ κ°μκ°€ λ§λ”μ§€ ν™•μΈ)
    if len(indices) == len(pred_values):
        for i, idx in enumerate(indices):
            final_predictions[idx] = pred_values[i]
            prediction_counts[idx] += 1
    else:
        print(f"   β οΈ ν¬κΈ° λ¶μΌμΉ: μΈλ±μ¤ {len(indices)}κ° vs μμΈ΅κ°’ {len(pred_values)}κ°")
        # ν¬κΈ°κ°€ λ‹¤λ¥΄λ©΄ μµμ† κ°μλ§νΌλ§ ν• λ‹Ή
        min_len = min(len(indices), len(pred_values))
        for i in range(min_len):
            final_predictions[indices[i]] = pred_values[i]
            prediction_counts[indices[i]] += 1

# μμΈ΅λμ§€ μ•μ€ λ°μ΄ν„° ν™•μΈ
unassigned_count = np.sum(prediction_counts == 0)
if unassigned_count > 0:
    print(f"β οΈ μμΈ΅λμ§€ μ•μ€ λ°μ΄ν„°: {unassigned_count}κ° (0μΌλ΅ μ μ§€)")

# μ¤‘λ³µ μμΈ΅ ν™•μΈ
duplicate_count = np.sum(prediction_counts > 1)
if duplicate_count > 0:
    print(f"β οΈ μ¤‘λ³µ μμΈ΅λ λ°μ΄ν„°: {duplicate_count}κ°")

print(f"\nβ… μμΈ΅ κ²°κ³Ό ν†µν•© μ™„λ£")
print(f"   π“ μ΄ μμΈ΅ κ°μ: {len(final_predictions):,}κ°")
print(f"   π“ μμΈ΅κ°’ ν†µκ³„: ν‰κ· ={np.mean(final_predictions):.2f}, μµλ€={np.max(final_predictions):.2f}")
```

## 9οΈβƒ£ μµμΆ… κ²°κ³Ό μ €μ¥ λ° ν‰κ°€

```python
# μµμΆ… κ²°κ³Όλ¥Ό test_dfμ— μ¶”κ°€
print("π’Ύ μµμΆ… κ²°κ³Ό μ €μ¥...")

# μ›λ³Έ test_df λ³µμ‚¬
result_df = test_df.copy()

# μμΈ΅ κ²°κ³Ό μ¶”κ°€ (μμκ°’ μ κ±°)
result_df['pred_heat_demand'] = np.maximum(final_predictions, 0).round(1)

# CSV νμΌ μ €μ¥
output_filename = 'catboost_branch_season_predictions.csv'
result_df.to_csv(output_filename, index=False)

print(f"π“ κ²°κ³Ό νμΌ μ €μ¥: {output_filename}")

# =============================================================================
# RMSE μ¤‘μ‹¬ μ„±λ¥ ν‰κ°€ (μ‹¤μ κ°’μ΄ μλ” κ²½μ°)
# =============================================================================

if 'heat_demand' in test_df.columns:
    print(f"\nπ“ RMSE μ„±λ¥ ν‰κ°€")
    print("=" * 60)
    
    # μ „μ²΄ RMSE
    y_true = test_df['heat_demand'].values
    y_pred = result_df['pred_heat_demand'].values
    
    # μμλ‚ NaN κ°’ μ κ±°
    valid_mask = ~(np.isnan(y_true) | np.isnan(y_pred))
    y_true_clean = y_true[valid_mask]
    y_pred_clean = y_pred[valid_mask]
    
    overall_rmse = np.sqrt(mean_squared_error(y_true_clean, y_pred_clean))
    overall_mae = mean_absolute_error(y_true_clean, y_pred_clean)
    correlation = np.corrcoef(y_true_clean, y_pred_clean)[0, 1]
    
    print(f"π† μ „μ²΄ μ„±λ¥:")
    print(f"   RMSE: {overall_rmse:.4f}")
    print(f"   MAE:  {overall_mae:.4f}")
    print(f"   μƒκ΄€κ³„μ: {correlation:.4f}")
    print(f"   μ ν¨ λ°μ΄ν„°: {len(y_true_clean):,}κ°")
    
    # μ‹μ¦λ³„ RMSE (ν•µμ‹¬!)
    print(f"\nπ“ μ‹μ¦λ³„ RMSE μ„±λ¥:")
    season_names = {0: 'λΉ„λ‚λ°©μ‹μ¦', 1: 'λ‚λ°©μ‹μ¦'}
    season_results = {}
    
    for season in [0, 1]:
        mask = (test_df['heating_season'] == season) & valid_mask
        if np.sum(mask) > 0:
            season_rmse = np.sqrt(mean_squared_error(y_true[mask], y_pred[mask]))
            season_mae = mean_absolute_error(y_true[mask], y_pred[mask])
            season_corr = np.corrcoef(y_true[mask], y_pred[mask])[0, 1] if np.sum(mask) > 1 else 0
            season_results[season] = {
                'rmse': season_rmse, 
                'mae': season_mae, 
                'corr': season_corr,
                'count': np.sum(mask)
            }
            
            print(f"   {season_names[season]:8s}: RMSE={season_rmse:7.4f} | MAE={season_mae:7.4f} | μƒκ΄€={season_corr:6.3f} | {np.sum(mask):,}κ°")
    
    # λΈλμΉλ³„ RMSE (μƒμ„/ν•μ„ λ¶„μ„)
    print(f"\nπ“ λΈλμΉλ³„ RMSE μ„±λ¥:")
    branch_results = {}
    
    for branch in sorted(test_df['branch_id'].unique()):
        mask = (test_df['branch_id'] == branch) & valid_mask
        if np.sum(mask) > 1:  # μµμ† 2κ° μ΄μƒμ λ°μ΄ν„°κ°€ μμ–΄μ•Ό RMSE κ³„μ‚° κ°€λ¥
            branch_rmse = np.sqrt(mean_squared_error(y_true[mask], y_pred[mask]))
            branch_mae = mean_absolute_error(y_true[mask], y_pred[mask])
            branch_results[branch] = {
                'rmse': branch_rmse,
                'mae': branch_mae, 
                'count': np.sum(mask)
            }
    
    if branch_results:
        # RMSE κΈ°μ¤€ μ •λ ¬
        sorted_branches = sorted(branch_results.items(), key=lambda x: x[1]['rmse'])
        
        print(f"   π¥‡ RMSE μ°μ λΈλμΉ (Top 5):")
        for i, (branch, metrics) in enumerate(sorted_branches[:5], 1):
            print(f"      {i}. λΈλμΉ {branch}: RMSE={metrics['rmse']:7.4f} | MAE={metrics['mae']:7.4f} | {metrics['count']:,}κ°")
        
        print(f"   π¥‰ RMSE κ°μ„  ν•„μ” λΈλμΉ (Bottom 5):")
        for i, (branch, metrics) in enumerate(sorted_branches[-5:], 1):
            print(f"      {i}. λΈλμΉ {branch}: RMSE={metrics['rmse']:7.4f} | MAE={metrics['mae']:7.4f} | {metrics['count']:,}κ°")
        
        # λΈλμΉλ³„ μ„±λ¥ ν†µκ³„
        rmse_values = [v['rmse'] for v in branch_results.values()]
        print(f"\n   π“ λΈλμΉλ³„ RMSE ν†µκ³„:")
        print(f"      ν‰κ· : {np.mean(rmse_values):.4f}")
        print(f"      ν‘μ¤€νΈμ°¨: {np.std(rmse_values):.4f}")
        print(f"      μµμ†: {np.min(rmse_values):.4f}")
        print(f"      μµλ€: {np.max(rmse_values):.4f}")
    
    # μ‹μ¦Γ—λΈλμΉ μ΅°ν•©λ³„ RMSE (μƒμ„ 10κ°λ§)
    print(f"\nπ”¥ μ‹μ¦Γ—λΈλμΉ μ΅°ν•©λ³„ RMSE (Ranking):")
    combo_results = []
    
    for season in [0, 1]:
        for branch in test_df['branch_id'].unique():
            mask = (test_df['heating_season'] == season) & (test_df['branch_id'] == branch) & valid_mask
            if np.sum(mask) > 1:
                combo_rmse = np.sqrt(mean_squared_error(y_true[mask], y_pred[mask]))
                combo_name = f"{season_names[season]}_{branch}"
                combo_results.append((combo_name, combo_rmse, np.sum(mask)))
    
    # RMSE κΈ°μ¤€ μ •λ ¬ν•μ—¬ ν‘μ‹
    combo_results.sort(key=lambda x: x[1])
    for i, (combo_name, rmse, count) in enumerate(combo_results, 1):
        print(f"   {i:2d}. {combo_name:15s}: RMSE={rmse:7.4f} | {count:,}κ°")
    
    # ν›λ ¨λ λ¨λΈλ“¤μ μµμ ν™” μ„±λ¥κ³Ό μ‹¤μ  ν…μ¤νΈ μ„±λ¥ λΉ„κµ
    print(f"\nπ” λ¨λΈ μµμ ν™” vs μ‹¤μ  μ„±λ¥ λΉ„κµ:")
    optimization_rmses = [v['best_score'] for v in training_results.values() if v.get('best_score') is not None]
    
    if optimization_rmses:
        print(f"   ν›λ ¨μ‹ μµμ ν™” RMSE: ν‰κ· ={np.mean(optimization_rmses):.4f}, λ²”μ„=[{np.min(optimization_rmses):.4f}, {np.max(optimization_rmses):.4f}]")
        print(f"   μ‹¤μ  ν…μ¤νΈ RMSE: {overall_rmse:.4f}")
        print(f"   μ„±λ¥ μ°¨μ΄: {abs(overall_rmse - np.mean(optimization_rmses)):.4f}")
    
else:
    print(f"\nβ οΈ ν…μ¤νΈ λ°μ΄ν„°μ— heat_demand μ»¬λΌμ΄ μ—†μ–΄μ„ RMSE ν‰κ°€λ¥Ό μν–‰ν•  μ μ—†μµλ‹λ‹¤.")
    print(f"   μμΈ΅ κ²°κ³Όλ§ μ €μ¥λμ—μµλ‹λ‹¤.")
    
    # μμΈ΅κ°’ κΈ°λ³Έ ν†µκ³„
    print(f"\nπ“ μμΈ΅κ°’ κΈ°λ³Έ ν†µκ³„:")
    print(f"   κ°μ: {len(result_df):,}κ°")
    print(f"   ν‰κ· : {result_df['pred_heat_demand'].mean():.2f}")
    print(f"   μ¤‘μ•™κ°’: {result_df['pred_heat_demand'].median():.2f}")
    print(f"   ν‘μ¤€νΈμ°¨: {result_df['pred_heat_demand'].std():.2f}")
    print(f"   λ²”μ„: [{result_df['pred_heat_demand'].min():.2f}, {result_df['pred_heat_demand'].max():.2f}]")
    
    # μ‹μ¦λ³„ μμΈ΅ ν†µκ³„
    print(f"\nπ“ μ‹μ¦λ³„ μμΈ΅ ν†µκ³„:")
    for season in [0, 1]:
        season_data = result_df[result_df['heating_season'] == season]['pred_heat_demand']
        if len(season_data) > 0:
            season_name = 'λΉ„λ‚λ°©μ‹μ¦' if season == 0 else 'λ‚λ°©μ‹μ¦'
            print(f"   {season_name}: ν‰κ· ={season_data.mean():.2f}, κ°μ={len(season_data):,}κ°")

print("=" * 60)
```
